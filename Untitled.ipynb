# SWIR trace elements

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
import pandas as pd
```

```python
%matplotlib inline
```

```python
# REE order for plots
elements = array([1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 12, 14])
el_names = ['La', 'Ce', 'Pr', 'Nd', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Yb']
chon = array([0.2347, 0.6032, 0.0891, 0.4524, 0.1471, 0.056, 0.1966, 0.0363, 0.2427, 0.0566, 0.1589, 0.1625])
```

```python
# Load file
xl = pd.ExcelFile('SWIR REEs.xlsx')

# Load sheet as numpy array
REE_data = xl.parse('REEs').as_matrix()

# Variables from input file
Sample = REE_data[:,0]
Group = REE_data[:,1]
REEs = REE_data[:,2:14]
Cr_num = REE_data[:,14]
fO2 = REE_data[:,2:15]
TiO2 = REE_data[:,2:16]
Lat = REE_data[:,2:17]
Long = REE_data[:,2:18]

# background fields
#Cr_num_ridge = xlsread('background fields.xlsx', 'Sheet1', 'F3:F190');
#Yb_ridge = xlsread('background fields.xlsx', 'Sheet1', 'B3:B190');

#Cr_num_veined = xlsread('background fields.xlsx', 'Sheet1', 'H3:H85');
#Yb_veined = xlsread('background fields.xlsx', 'Sheet1', 'I3:I85');
```

```python
# Make REE figure for each group

fig, axes = plt.subplots(nrows=1, ncols=4,figsize=(12,3))
#axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

for i in range(len(REEs)):
    if Group[i] in [1,11]:# == 1 or REE_data.ix[i,'Group'] == 11:
        sp = 2
    elif Group[i] in [2,3]:# == 2 or REE_data.ix[i,'Group'] == 3:
        sp = 1
    elif Group[i] in [4,5,6,9,10]:#== 4 or REE_data.ix[i,'Group'] == 5: #or group(i) == 6 or group(i) == 9 or group(i) == 10:
        sp = 0
    elif Group[i] in [7,8]:#== 7 or REE_data.ix[i,'Group'] == 8: 
        sp = 3

    axes[sp].plot(elements, REEs[i]/chon, 'r', marker='o')
    
for i in range(4): 
    axes[i].set_yscale('log')   
    axes[i].set_ylim([10e-6, 10e1])
    axes[i].set_xlim([1, 14])
    axes[i].set_xticks(elements)
    axes[i].set_xticklabels(el_names, fontsize=10)
    axes[i].set_yticks(logspace(-5,2,8))
    axes[i].set_ylabel('Conc/Chon')
    axes[i].set_title('SWIR REEs')
    fig.tight_layout()

plt.savefig("test.eps", format="eps")
```
